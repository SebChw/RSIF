```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
OKABE_ITO_SCALE = ["#E69F00", "#56B4E9", "#009E73", "#F0E442", "#0072B2", "#D55E00", "#CC79A7"]

%load_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv('default_IForest_ISF_RISF.csv')
g = sns.FacetGrid(df, col="clf")
g.map(sns.barplot,'dataset_name', 'auc')
plt.gcf().set_size_inches(15, 8) # We don't need to check Isolation forest, maybe just for debugging.
```

```python
results = pd.read_csv("results_max_samples.csv")

sns.set_palette(sns.color_palette(OKABE_ITO_SCALE))
g = sns.relplot(data=results, x="max_samples", y="auc", hue="clf", marker="o", col="dataset_name", col_wrap=4, 
            facet_kws={'sharey': False, 'sharex': False}, kind="line")

g.fig.suptitle('INFLUENCE OF MAX_SAMPLES')
plt.gcf().set_size_inches(20, 10)
```
